```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

#### Functional Method

```python
# plt.subplot(row, column, object_number)
plt.plot(x,y)

plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.show()
```

```python
# subplot(row, column, object_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

#### Object Oriented

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2 = fig.add_axes([0.1,0.5, 0.4, 0.4])
axes2.plot(y,x)
axes2.set_title('SMALLer PLOT')
```
