```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

#### Functional Method

```python
# plt.subplot(row, column, object_number)
plt.plot(x,y)

plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.show()
```

```python
# subplot(row, column, object_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

#### Object Oriented

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2 = fig.add_axes([0.1,0.5, 0.4, 0.4])
axes2.plot(y,x)
axes2.set_title('SMALLer PLOT')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# for current_axes in axes:
#     current_axes.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[0].set_xlabel('X Label')
axes[0].set_ylabel('Y Label')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
axes[1].set_xlabel('Y Label')
axes[1].set_ylabel('X Label')

# fix layouts
plt.tight_layout()
```

#### Figure Size, Aspect Ratio & DPI

```python
fig = plt.figure(figsize=(5,2), dpi=250)

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2), dpi=120)

axes[0].plot(x, y)
axes[1].plot(y,x)

plt.tight_layout()

# save figure
fig.savefig('exaple.png', dpi=200)
```

#### Legend

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Square')
ax.plot(x, x**3, label='X Cube')

ax.legend()
fig.savefig('exaple.png', dpi=200)
```

#### Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, 
        color='red', 
        linewidth=2, 
        alpha=0.5, 
        linestyle='-', 
        marker='1', 
        markersize=12,
        markerfacecolor='yellow',
        markeredgewidth=3,
        markeredgecolor='green'
       )
```

#### Plot Range

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
